# concatenar,mesclar e juntar

São três maneiras de combinar dataframes

```python
df3=pd.DataFrame({'a': (1,2,3),'b':(3,6,7)})
df4=pd.DataFrame({'c':(78,4,8),'d':(9,2,47)})
```

concatenar-basicamente colar dataframe,mas dimensões tem que corresponder ao longo do eixo que você esta concatenando

```python
df5=pd.concat([df3,df4])
df5
```

juntar-É um método para combinar colunas de dois dataframes diferentes

```python
df6=df3.join(df4)
df6
```

mesclar-A função ** mesclar ** permite que você mescle os quadros de dados juntos usando uma lógica semelhante à mesclagem de tabelas SQL juntas

```python
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
pd.merge(esquerda,direita,how='inner',on='key')
```
